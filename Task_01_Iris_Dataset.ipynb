# Iris Dataset – Data Exploration & Visualization

This notebook explores the Iris dataset to understand its structure, distribution,
and relationships between features using data visualization techniques.
The goal is to practice data inspection and visual analysis, which are essential
steps before applying any machine learning model.


```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Load the uploaded CSV file
df = pd.read_csv("iris.csv")

# Display first 5 rows
df.head()
```

```python
# Shape of dataset
df.shape
```

```python
# Column names
df.columns
```

```python
df.info()
```

```python
# Checking missing values
df.isnull().sum()
```

```python
df.head()
```

```python
df['species'].value_counts()

plt.figure(figsize=(6,4))
sns.countplot(x='species', data=df)
plt.title("Distribution of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x='petal_length',
    y='petal_width',
    hue='species',
    data=df
)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['sepal_length'], bins=20)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title("Petal Length by Species")
plt.show()
```

## Real-World Consideration

If this dataset were collected from real biological measurements,
small measurement errors in petal length could significantly impact
classification decisions. Petal dimensions appear to be the most
influential features.

##Key Observations
- Dataset contains balanced species distribution.
- Petal features are highly correlated and strongly differentiate species.
- Setosa is clearly separable from other species.
- Some overlap exists between Versicolor and Virginica.


##Conclusion
